```python
# 1.1 Visualization - Matrix with ages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fields = ['iid', 'age', 'pid', 'dec']
speed_date = pd.read_csv('speed_dating_assignment.csv', usecols=fields)

#remove unnecessary columns
speed_date = speed_date[['iid', 'age', 'pid', 'dec']]

#remove rows with missing data
speed_date = speed_date[speed_date.age == speed_date.age]
speed_date = speed_date[speed_date.pid == speed_date.pid]
speed_date = speed_date.reset_index(drop=True)

#get new length of matrix
length = len(speed_date)
print(length)
```

```python
#get minimum and maximum age in Matrix
minage = 100
maxage = 0
for y in range(0, length):
    age = int(speed_date.iloc[y]['age'])
    if age > maxage:
        maxage = age
    if age < minage:
        minage = age
```

```python
#create two new matrices, matrix1 is for total decisions, matrix2 is for frequency
index = list()
columns = list()
for age in range(minage, maxage+1):
    index.append(age)
    columns.append(age)
matrix1 = pd.DataFrame(index=index, columns=columns, dtype='float64')
matrix1 = matrix1.fillna(0)
matrix2 = matrix1.copy(deep=True)
```

```python
#fill matrices with data from speed_date
for y in range(0, length):
    iid = int(speed_date.iloc[y]['iid'])
    age = int(speed_date.iloc[y]['age'])
    pid = int(speed_date.iloc[y]['pid'])
    series = speed_date.index[speed_date['iid'] == pid]
    #print(series)
    if series.size != 0:
        index = series[0]
        #print("index: " + str(index))
        partnerage = int(speed_date.iloc[index]['age'])
        dec = int(speed_date.iloc[y]['dec'])
        #print("iid: "+ str(iid) + ", age: "+ str(age) + ", pid: " + str(pid) + ", partnerage: " + str(partnerage) + ", dec: " + str(dec))
        matrix1.iat[age-minage, partnerage-minage] += dec
        matrix2.iat[age-minage, partnerage-minage] += 1
#print(matrix1)
#print(matrix2)
```

```python
#compute average decision
for y in range(0, maxage-minage+1):
    for x in range(0, maxage-minage+1):
        if matrix2.iat[y, x] != 0:
            matrix1.iat[y, x] /= matrix2.iat[y, x]
```
